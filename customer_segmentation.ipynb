```python
import numpy as mp
import pandas as pd
import os
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
df= pd.read_csv("supermarket_sales.csv", sep=",")
df=df.fillna(0)
df["Gender"]= df["Gender"].replace("Female",1) 
df["Gender"]= df["Gender"].replace("Male",0) 
km = KMeans(n_clusters=5)
x=df.iloc[:,[2,3,16]]
clusters = km.fit_predict(x)
df["label"] = clusters
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.Customer_age[df.label == 0], df["monthly_income"][df.label == 0], df["Gender"][df.label == 0], c='blue', s=60)
ax.scatter(df.Customer_age[df.label == 1], df["monthly_income"][df.label == 1], df["Gender"][df.label == 1], c='red', s=60)
ax.scatter(df.Customer_age[df.label == 2], df["monthly_income"][df.label == 2], df["Gender"][df.label == 2], c='green', s=60)
ax.scatter(df.Customer_age[df.label == 3], df["monthly_income"][df.label == 3], df["Gender"][df.label == 3], c='orange', s=60)
ax.scatter(df.Customer_age[df.label == 4], df["monthly_income"][df.label == 4], df["Gender"][df.label == 4], c='purple', s=60)
ax.view_init(30, 185)
plt.xlabel("Age")
plt.ylabel("Monthly Income")
ax.set_zlabel('Gender')
plt.show()
```
